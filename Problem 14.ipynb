```python
# Import Libraries

import pandas as pd
import re

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Complaint-Response Dataset

data = {
    "Complaint": [
        "My bill is too high this month",
        "Internet is not working",
        "App keeps crashing",
        "Payment deducted twice",
        "Support team is not responding"
    ],
    
    "Response": [
        "We apologize for the inconvenience. Please check your bill breakdown or contact billing support.",
        
        "We regret the inconvenience. Please restart your router and contact technical support if issue persists.",
        
        "We apologize for the inconvenience. Please reinstall the app or update to the latest version.",
        
        "We apologize for the inconvenience. Refund will be processed after verification.",
        
        "We apologize for the delay. Our support team will contact you shortly."
    ]
}

df = pd.DataFrame(data)

display(df)
```

```python
# Text Preprocessing

def preprocess(text):
    
    text = text.lower()
    
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    return text
```

```python
# Preprocess Complaints

df["Cleaned Complaint"] = df["Complaint"].apply(preprocess)
```

```python
# TF-IDF Vectorization

vectorizer = TfidfVectorizer(
    stop_words='english'
)

X = vectorizer.fit_transform(
    df["Cleaned Complaint"]
)
```

```python
# User Complaint Input

user_complaint = input("Enter Customer Complaint: ")

cleaned_input = preprocess(user_complaint)

input_vec = vectorizer.transform([cleaned_input])
```

```python
# Similarity Calculation

similarities = cosine_similarity(
    input_vec,
    X
)

best_match_index = similarities.argmax()
```

```python
# Retrieve Recommended Response

recommended_response = df.iloc[
    best_match_index
]["Response"]

matched_complaint = df.iloc[
    best_match_index
]["Complaint"]

similarity_score = similarities[0][best_match_index]

print("Input Complaint:", user_complaint)
print("\nMost Similar Past Complaint:", matched_complaint)
print("Similarity Score:", round(similarity_score, 2))
print("\nRecommended Response:")
print(recommended_response)
```
